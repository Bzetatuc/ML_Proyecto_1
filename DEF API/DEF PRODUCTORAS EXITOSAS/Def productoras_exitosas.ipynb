```python
import locale
#m√≥dulo locale en Python proporciona funciones para realizar operaciones de formato y manipulaci√≥n de datos basadas en las configuraciones locales del sistema operativo.


# Configurar el locale para el formato de moneda en d√≥lares
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
```

```python
import locale

def productoras_exitosas(Productora: str):
    productora_lower = Productora.lower()

    # Rellenar los valores faltantes (NaN) en la columna 'name_companies' con una cadena vac√≠a ('')
    df_Prod_exitosas['name_companies'] = df_Prod_exitosas['name_companies'].fillna('')

    # Filas correspondientes a la productora
    peliculas_productora = df_Prod_exitosas[df_Prod_exitosas['name_companies'].str.lower().str.contains(productora_lower)]

    # Total de revenue
    total_revenue = peliculas_productora['revenue'].sum()

    # Cantidad de pel√≠culas realizadas por la productora
    cantidad_peliculas = len(peliculas_productora)

    # Formatear el revenue como moneda en d√≥lares
    revenue_formateado = locale.currency(total_revenue, grouping=True)

    Resultado = f"La productora {Productora} ha tenido un revenue de u$s {revenue_formateado} y ha realizado {cantidad_peliculas} pel√≠culas"
    return Resultado

```
