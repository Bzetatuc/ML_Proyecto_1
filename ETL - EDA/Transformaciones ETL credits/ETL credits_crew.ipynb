```python
df = pd.read_csv("//Users/benjaminzelaya/Desktop/dataProyecto/credits.csv")
df.head(3)
```

```python
# Obtener las columnas actuales del DataFrame
columnas_actuales = df.columns.tolist()

# Definir el nuevo orden de las columnas
nuevo_orden = ['id', 'cast', 'crew']

# Reordenar las columnas del DataFrame
df = df[nuevo_orden]
```

```python
import ast 

df['crew'] = df['crew'].apply(ast.literal_eval)
```

```python
# Desanidar la columna production_countries y crear nuevas columnas
df_exploded = df.explode('crew')

def extract_country_info(x):
    if isinstance(x, dict):
        return pd.Series([x.get('credit_id', ''), x.get('department', ''),x.get('gender', ''),x.get('id', ''),x.get('job', ''),x.get('name', ''),x.get('profile_path', '')  ])
    else:
        return pd.Series(['', ''])

df_exploded[['credit_id', 'department', 'gender','id','job', 'name', 'profile_path']] = df_exploded['crew'].apply(extract_country_info)

```

```python
df = df.join(df_exploded[['credit_id', 'department', 'gender', 'id', 'job', 'name', 'profile_path']], lsuffix='_df', rsuffix='_df_exploded')
```
